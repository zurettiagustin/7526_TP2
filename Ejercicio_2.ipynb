```python
import random as rn
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sim():

    # const

    time = 1000000 # 1000 segundos
    interval = 10 # 10 ms
    p = 1/40
    q = 1/30

    # vars
    
    status = 0
    total = 0
    ranges = []
    statuses = []
    free = 0

    for x in range(0, time, interval):

        rnd = rn.random()
        pq = (p * (1-q))
        qp = (q *  (1-p))
        
        if (status == 0):
            if (rnd <= pq):
                total += 1
                status += 1
            else:
                free += interval
        else:
            if (rnd <= qp):
                status -= 1
            elif (rnd >= 1 - pq):
                total += 1
                status += 1
        
        
        ranges.append(x)
        statuses.append(status)

    return ranges, statuses, total, free

ranges, statuses, total, free = sim()

print("b)")

plt.figure(figsize=(12,10))
plt.plot(ranges, statuses)
plt.xlabel('Instante')
plt.ylabel('Cantidad')
```

```python
print("c)")
plt.figure(figsize=(12,10))
plt.hist(statuses)
plt.xlabel('Estados')
plt.ylabel('Cantidad')
plt.show()
```

```python
print("d)")

free_seconds = (free / 1000)
free_proportion = ((free_seconds * 100) / 1000)

print(str(free_proportion) + "%")
```
