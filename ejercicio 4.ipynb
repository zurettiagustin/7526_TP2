```python
import matplotlib.pyplot as plt

# Condiciones iniciales

tiempo = 145

Capacidad_Maxima_salud = 0.3

poblacion_recuperada = 0

poblacion_susceptible = 0.97

poblacion_infectada = 0.03

beta = 0.27

gamma = 0.043

Capacidad_Maxima_salud = 0.3
```

```python
# Implementacion del modelo SIR.

def modelar_susceptibles(susceptibles, infectados):
    return susceptibles - susceptibles * infectados * beta

def modelar_infectados(susceptibles, infectados):
    return beta * susceptibles * infectados + infectados - gamma * infectados

def modelar_recuperados(recuperados, infectados):
    return gamma * infectados + recuperados 


# Realizo la simulacin de la pandemia.
lista_susceptibles = [poblacion_susceptible]
lista_infectados = [poblacion_infectada]
lista_recuperados = [poblacion_recuperada]
sistemaSalud = [ Capacidad_Maxima_salud ]


for i in range(1, tiempo):
    poblacion_susceptible = modelar_susceptibles(poblacion_susceptible, poblacion_infectada)
    lista_susceptibles.append(poblacion_susceptible)
        
    poblacion_infectada = modelar_infectados(poblacion_susceptible, poblacion_infectada)
    lista_infectados.append(poblacion_infectada)
    
    poblacion_recuperada = modelar_recuperados(poblacion_recuperada, poblacion_infectada)
    lista_recuperados.append(poblacion_recuperada)
    
    sistemaSalud.append(Capacidad_Maxima_salud)

for i in range(len(lista_infectados)):
        if lista_infectados[i] < float(0.001) and i > 100 :
            print(i)
            break
```

```python
plt.figure(figsize=(12, 8))
plt.title('Evolucin de la pandemia en 145 das')
plt.plot(lista_susceptibles)
plt.plot(lista_infectados)
plt.plot(lista_recuperados)
plt.plot(sistemaSalud)
plt.legend(['Susceptible', 'Infectados', 'Recuperados', 'Lmite de asistencias del Sistema de Salud'])
plt.xlabel('Das')
plt.ylabel('Fraccin de la Poblacin')
plt.show()
```
